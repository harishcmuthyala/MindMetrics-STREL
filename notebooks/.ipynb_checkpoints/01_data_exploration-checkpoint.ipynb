Goals:

```python
import pandas as pd
import numpy as np
import yaml
import matplotlib.pyplot as plt
import seaborn as sns

# â”€â”€ load central config â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
with open('config/config.yaml') as f:
    cfg = yaml.safe_load(f)

print('Config loaded:', list(cfg.keys()))
```

```python
# â”€â”€ load raw data â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv(cfg['paths']['raw_data'])

print('Shape :', df.shape)
print('Columns:', df.columns.tolist())
df.head()
```

```python
# â”€â”€ basic info & dtypes â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df.info()
```

```python
# â”€â”€ missing values â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
# â”€â”€ missing-value heatmap â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import seaborn as sns
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='YlOrRd')
plt.title('Missing Values')
plt.tight_layout()
plt.show()
```

```python
# â”€â”€ class balance of target labels â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
for col in ['NHR_Stress', 'SNS_Stress']:
    print(f'\n{col}:')
    print(df[col].value_counts())
    print(df[col].value_counts(normalize=True).round(3))
```

```python
# â”€â”€ descriptive statistics (numerical columns) â”€â”€
df.describe().T
```

```python
# â”€â”€ distribution plots â€” physiological signals â”€â”€
physio_cols = cfg['features']['physiological']

fig, axes = plt.subplots(len(physio_cols), 1, figsize=(10, 3*len(physio_cols)))
for ax, col in zip(axes, physio_cols):
    df[col].dropna().hist(bins=50, ax=ax, color='steelblue', edgecolor='white')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
# â”€â”€ correlation heatmap (numerical features) â”€â”€â”€â”€â”€
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

plt.figure(figsize=(14, 10))
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# â”€â”€ categorical breakdowns â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
for col in cfg['features']['categorical']:
    print(f'\n{col}:')
    print(df[col].value_counts())
```

```python
# â”€â”€ unique participants â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print('Unique participants:', df['Participant'].nunique())
print(sorted(df['Participant'].unique()))
```
